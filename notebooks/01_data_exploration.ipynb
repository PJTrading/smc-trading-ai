```python
# import pandas as pd
# from pathlib import Path

# # Definir la ruta al archivo
# file_path = Path.cwd() / "data" / "raw" / "ES=F_data_1h.csv"

# # Cargar los datos
# # df = pd.read_csv(file_path, index_col='Datetime', parse_dates=True)
# df = pd.read_csv(file_path, index_col=0, parse_dates=True)

# # Mostrar información y las primeras filas
# print(df.info())
# display(df.head())

import pandas as pd
from pathlib import Path

# Definir la ruta al archivo
file_path = Path.cwd() / "data" / "raw" / "ES=F_data_1h.csv"

# --- PASO 1: Leer el CSV correctamente ---
# Le decimos a pandas que las primeras dos filas (0 y 1) son el encabezado.
df = pd.read_csv(file_path, header=[0, 1], index_col=0, parse_dates=True)

# --- PASO 2: Simplificar el encabezado ---
# El encabezado actual es un "MultiIndex". Vamos a simplificarlo.
# Nos quedamos solo con el segundo nivel del encabezado ('Open', 'High', etc.)
df.columns = df.columns.get_level_values(1)

# Renombramos el índice para que tenga un nombre limpio
df.index.name = 'Datetime'

# --- PASO 3: Verificar el resultado ---
# Ahora la salida debería ser correcta.
print("--- Información del DataFrame Corregido ---")
print(df.info())

print("\n--- Primeras 5 filas del DataFrame Corregido ---")
display(df.head())
```
